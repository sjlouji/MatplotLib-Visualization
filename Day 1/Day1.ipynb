```python
import matplotlib.pyplot as p
import numpy as np
```

```python
population_in = np.arange(1000,4000, 600)
population_us = [2000,3000,1000,4000, 9000]
print(population_us)
```

```python
year_in = np.arange(2000,2005)
year_us = np.arange(2000,2005)
print(year_us)
```

```python
p.bar(year_in,population_in, alpha = .5, linewidth= 10, width = .9, label="India")
p.bar(year_us,population_us, alpha = .5, linewidth= 10, width = .9, label="United States")
p.legend()
```
