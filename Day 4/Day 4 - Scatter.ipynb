# Scatter Plots

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
x = [5,2,6,2,7,5,8,9,3,2,1,7,8,9,2,7,8,2,8,2,7]
y = [2,5,6,7,8,1,2,3,4,6,7,5,3,2,5,6,3,8,4,6,2]

colors = [7,8,3,2,4,5,6,7,7,2,3,5,6,7,7,2,3,4,5,6,3]
```

```python
plt.style.use('seaborn')
plt.scatter(x,y, c=colors, s=60 ,edgecolor="black", cmap='Greens')
cbar = plt.colorbar()
cbar.set_label('Satisfaction')
```

Example

```python
fs = pd.read_csv('scatter.csv')
print(fs.head())
```

```python
plt.scatter(fs['view_count'],fs['likes'], edgecolor='black',c=fs['ratio'],s=80, cmap='summer')
cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('View Count')
plt.ylabel('Like Count')
```
