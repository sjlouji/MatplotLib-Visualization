```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
fs = pd.read_csv('data.csv')
```

```python
age = fs['Age']
dev_sal = fs['All_Devs']
python_sal = fs['Python']
javascript_sal = fs['JavaScript']
```

```python
plt.plot(age,dev_sal, color='orange', linestyle='--', label="All Developers")
plt.plot(age,python_sal, color='green', label="Python Developers")
plt.plot(age,javascript_sal, color='yellow', label="Python Developers")
plt.fill_between(age,python_sal, dev_sal,alpha=.2, where = (python_sal > dev_sal) ,interpolate=True,color="green")
plt.fill_between(age,python_sal, dev_sal, where = (python_sal <= dev_sal) ,interpolate=True,color="orange")
plt.fill_between(age,python_sal, dev_sal, where = (python_sal <= javascript_sal) ,interpolate=True,color="yellow")

plt.legend()
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary for diffrent Developers')
plt.tight_layout()
```
