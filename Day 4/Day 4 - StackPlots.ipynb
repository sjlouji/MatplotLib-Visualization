# StackPlots

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = [1,2,3,4,5,6,7]
jocking = [0.5,1,1,1.5,1.5,2,2]
skipping = [1,1,1,.5,.5,1,2]
excersice = [0,.3,.3,1,1.5,1.5,1]
```

```python
plt.stackplot(days,jocking, skipping, excersice, labels=['Jocking','Skipping','Ground Excersice'])
plt.legend(loc='upper left')
plt.title('Day vs excersice')
```
