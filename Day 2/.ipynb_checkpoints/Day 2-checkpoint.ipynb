# Line

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
age = [25,30,40,45,47,48,50,60]
devope_sal = [4312,5432,6543,7645,8765,9568,10034,11345]
```

```python
plt.plot(age, devope_sal)
plt.title('Average Salary(EUR) by Age')
plt.xlabel('Age')
plt.ylabel('Salary (EUR)')
```

```python
sd_sal = [1234,1246,2364,5634,6334,7234,10345,11498]
plt.style.use('default')
plt.plot(age, sd_sal, 'k--', )
plt.plot(age, devope_sal, 'b', marker=".")
plt.legend(['Software Developer Salary','DevOps Engineer Salary'])
```

```python
js_sal = [1231,1325,1534,2334,3234,4465,5234,6034]
plt.style.use('default')
plt.plot(age, sd_sal, 'k--', )
plt.plot(age, devope_sal, 'b', marker=".")
plt.plot(age, js_sal, 'r', )
plt.grid()
plt.tight_layout()
plt.savefig('first.png')
```

# Bar

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
age = [25,30,35,40,45,50,55,60]
pydev_sal = [4000, 6000, 8000, 10000, 12000, 14000, 16000, 18000]
```

```python
plt.bar(age,pydev_sal, color="g" ,label="Python Developers")
plt.xlabel('Age')
plt.ylabel('salary')
plt.title('Age and Salary')
plt.legend()
```

```python
temp = np.arange(len(age))
```

```python
jsdev_salary = [4000, 3000, 5000, 4300, 8000, 9000, 8500, 10000]
erlangdev_salary = [10000, 8000, 7000, 10300, 12000, 14000, 15000, 18000]
```

```python
plt.bar(temp, jsdev_salary, label="Javascript Developers")
```

```python
width = .25
plt.bar(temp + width,pydev_sal, color="g" ,label="Python Developers", width=width)
plt.bar(temp, jsdev_salary, label="Javascript Developers", color="y" ,width=width)
plt.bar(temp - width, erlangdev_salary, label="Erlang Developers", color="r" ,width=width)
plt.xlabel('Age')
plt.ylabel('salary')
plt.title('Age and Salary')
plt.legend()
plt.xticks(ticks=temp, labels=age)
plt.grid()
plt.tight_layout()
```

Example

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
from collections import Counter
%matplotlib inline
```

```python
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    language_counter = Counter()
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
```

```python
languages = []
used = []
print(language_counter)
for val in language_counter:
    languages.append(val[0])
    used.append(val[1])
print(languages)
print(used)
```
