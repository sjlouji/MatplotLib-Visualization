```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
age = [25,30,40,45,47,48,50,60]
devope_sal = [4312,5432,6543,7645,8765,9568,1234,2345]
```

```python
plt.plot(age, devope_sal)
plt.title('Average Salary(EUR) by Age')
plt.xlabel('Age')
plt.ylabel('Salary (EUR)')
```

```python
sd_sal = [1234,546,2364,7634,6334,1234,6345,3498]
plt.plot(age, sd_sal, 'k--', marker="o")
plt.plot(age, devope_sal, 'b', marker="*")
plt.legend(['Software Developer Salary','DevOps Engineer Salary'])
```

```python
js_sal = [4231,125,6234,2734,1234,3465,1234,634]
plt.plot(age, sd_sal, 'k--', marker="o")
plt.plot(age, devope_sal, 'b', marker="*")
plt.plot(age, js_sal, 'r', marker="*")
```
