# PIE Charts

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
slice = [60,30,40,20]
labels = ['Python', 'Dart', 'JavaScript', 'Erlang']
colors=['blue','red','yellow','green']
explode = [0,0,0,0.1]
```

```python
plt.pie(slice, labels=labels, colors=colors, explode=explode, wedgeprops={'edgecolor':'white'})
plt.tight_layout()
plt.title('Language Pie chart')
```

Example without pandas

```python
import pandas as pd
import numpy as np
%matplotlib inline
import csv
from collections import Counter
```

```python
with open('data.csv') as file:
    file = csv.DictReader(file, delimiter=',')
    language_counter = Counter()
    for row in file:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
    print(language_counter)
```

```python
languages = []
used = []
for val in language_counter.most_common(10):
    languages.append(val[0])
    used.append(val[1])
print(languages)
print(used)
```

```python
plt.pie(used,labels=languages, autopct="%1.1f%%")
plt.tight_layout()
plt.title('Most used Languages')
```

Example with pandas

```python
val = pd.read_csv('data.csv')
languages = val['LanguagesWorkedWith']
used = val['Responder_id']
coun = Counter()
for data in languages:
    coun.update(data.split(';'))
language = []
used = []
for vals in coun.most_common(10):
    language.append(vals[0])
    used.append(vals[1])
plt.pie(used,labels=language, autopct="%1.1f%%")
plt.tight_layout()
plt.title('Most used Languages')
```
